```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n_success_observed = [10,15]  # Use the existing variable
n_draws = 60
n_success_pop = 100  # successes in population
def prior(N):
    return 1
pop_size = np.linspace(150, 2000, 100)  
likelihoods = []
for n_success in n_success_observed:
    likelihoods = []
    posterior = []    
    for M in pop_size:
        hyper_pmf = stats.hypergeom.pmf(
        k=n_success, 
        M=int(M), 
        n=n_success_pop, # successes in population 
        N=n_draws # sample size
        ) 
        likelihoods.append(hyper_pmf) 
        posterior.append(hyper_pmf * prior(M))

    plt.plot(pop_size, likelihoods, linewidth=2, label=f'Likelihood for {n_success} successes')
    plt.xlabel('Population Size (M)')
    plt.ylabel('P(X=k)')
    plt.title('Hypergeometric Likelihood')
    plt.grid(True, alpha=0.3)
    plt.plot(pop_size, posterior, linewidth=2, label=f'Posterior for {n_success} successes')
    plt.legend()
```

```python
n_success_observed = [10,15]  # Use the existing variable
n_draws = 60
n_success_pop = 100  # successes in population
def prior(N):
    return 1/N
pop_size = np.linspace(150, 2000, 100)  
likelihoods = []
for n_success in n_success_observed:
    likelihoods = []
    posterior = []    
    for M in pop_size:
        hyper_pmf = stats.hypergeom.pmf(
        k=n_success, 
        M=int(M), 
        n=n_success_pop, # successes in population 
        N=n_draws # sample size
        ) 
        likelihoods.append(hyper_pmf) 
        posterior.append(hyper_pmf * prior(M)*400)

    plt.plot(pop_size, likelihoods, linewidth=2, label=f'Likelihood for {n_success} successes')
    plt.xlabel('Population Size (M)')
    plt.ylabel('P(X=k)')
    plt.title('Hypergeometric Likelihood')
    plt.grid(True, alpha=0.3)
    plt.plot(pop_size, posterior, linewidth=2, label=f'Posterior for {n_success} successes')
    plt.legend()
    max_likelihood_index = np.argmax(likelihoods)
    max_likelihood_M = pop_size[max_likelihood_index]
    print(f"Maximum likelihood estimate of population size M: {max_likelihood_M:.2f}")

    max_posterior_index = np.argmax(posterior)
    max_posterior_M = pop_size[max_posterior_index]
    print(f"Maximum posterior estimate of population size M: {max_posterior_M:.2f}")
    



  
```

```python
V_fish = 10 
e_V_fish = 1
V_lake = 5000
e_V_lake = 300

N_fish = int(V_lake/V_fish)
e_N_fish = np.sqrt((e_V_lake/V_fish)**2 + (V_lake *e_V_fish/V_fish**2)**2)

```

```python
def gauss_prior(x, mu, sigma):
    return (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
```
