```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x, a, b):
    return 1 + a*x +b*x**2

x_axis = np.random.uniform(-1, 1,2000)
plt.plot(x_axis, function(x_axis, 0.5, 0.5), 'o', markersize=1)

```

```python
#generate random points from the dsitribtuion
samples = []
tot_samples = 0
while len(samples) < 2000:
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(0, 2.5)
    if y < function(x, 0.5, 0.5):
        samples.append(x)
    tot_samples += 1
```

```python
# save as .csv file

np.savetxt("samples.csv", samples, delimiter=",")
```

```python
Nbins= 20
x_range = (-1, 1)
print(f"Total samples generated: {tot_samples}")
counts, edges = np.histogram(samples, bins=Nbins, range=x_range, density=True)
bin_centers = (edges[:-1] + edges[1:]) / 2
plt.bar(bin_centers, counts, width=edges[1] - edges[0], alpha=0.5, label='Sampled Distribution')
x_axis = np.linspace(-1, 1, 1000)
plt.plot(x_axis, function(x_axis, 0.5, 0.5), 'r-', label='Target Distribution')
plt.legend()
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Rejection Sampling')
plt.show()
```

```python
from scipy import integrate

def function(x, a, b):
    return 1 + a*x + b*x**2

# Calculate the integral of the function over [-1, 1]
integral, _ = integrate.quad(lambda x: function(x, 0.5, 0.5), -1, 1)

def normalized_function(x, a, b):
    return function(x, a, b) / integral

# Now use the normalized function
x_axis = np.linspace(-1, 1, 1000)
plt.plot(x_axis, normalized_function(x_axis, 0.5, 0.5), 'r-', label='Target Distribution')
plt.bar(bin_centers, counts, width=edges[1] - edges[0], alpha=0.5, label='Sampled Distribution')
plt.legend()
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Rejection Sampling')
plt.show()
```
